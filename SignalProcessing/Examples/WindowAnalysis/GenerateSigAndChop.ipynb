```python
import numpy as np
import math
import matplotlib.pyplot as plt 
import pandas as pd
# import random
import os
```

# Visualize signal

## Pure signal

```python
# signal sinwave
# y = sin(wt) 
# y = sin(2*pi*f*t)

# Generate signal
fsig = 50 # Hz

# sampling rate 
fs=1000 # Hz
Ts=1/fs # Period
lenSignal = 0.1 # sec
# create time stamp
t = np.arange(0,lenSignal,Ts)

y = np.sin(2*math.pi*fsig*t)
plt.plot(t,y,'ko-')
plt.xlabel('sec')
```

## Add noise

```python
nPointSig = len(y)
noise = np.random.rand(nPointSig)
y = np.sin(2*math.pi*fsig*t) + noise
plt.plot(t,y,'ko-')
plt.xlabel('sec')
```
